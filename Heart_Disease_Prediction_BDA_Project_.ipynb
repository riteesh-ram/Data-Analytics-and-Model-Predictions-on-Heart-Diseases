**Logistic Regression**

```python
#input and output selection
ip=df.drop(['Heart_attack'],axis=1)
op=df['Heart_attack']
```

```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts=train_test_split(ip,op,test_size=0.3, random_state=3)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(xtr)
xtr=sc.transform(xtr)
xts=sc.transform(xts)
```

```python
xtr.shape
```

```python
xts.shape
```

```python
from sklearn.linear_model import LogisticRegression
alg=LogisticRegression()
```

```python
#train the algorithm with the training data
alg.fit(xtr,ytr)
yp1=alg.predict(xts)
```

```python
yp1
```

```python
from sklearn import metrics
import scikitplot as skplt
skplt.metrics.plot_confusion_matrix(yts ,yp1 ,figsize=(10,5))
```

```python
accuracy=metrics.accuracy_score(yts,yp1)
print(accuracy)
```

```python
from sklearn.metrics import classification_report
```

```python
log_report = classification_report(yp1, yts)
print(log_report)
```

**K-NN**

```python
#KNN algorithm the nearest distance is calculated
from sklearn.neighbors import KNeighborsClassifier

neighbors=np.arange(1,9)
train_accuracy=np.empty(len(neighbors))

for i,k in enumerate(neighbors):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(xtr,ytr)
    train_accuracy[i]=knn.score(xtr,ytr)

plt.xlabel('neighbors of number')
plt.ylabel('accuracy')
plt.title('k-NN Varying number of neighbors')
plt.plot(neighbors, train_accuracy, label='Training accuracy')
plt.legend()
plt.show()
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
knn.fit(xtr,ytr)
yp3=knn.predict(xts)
```

```python
yp3
```

```python
from sklearn import metrics
import scikitplot as skplt
skplt.metrics.plot_confusion_matrix(yts ,yp3 ,figsize=(10,5))
```

```python
accuracy=metrics.accuracy_score(yts,yp3)
print(accuracy)
```

```python
log_report = classification_report(yp3, yts)
print(log_report)
```

**Random Forest**

```python
xtr,xts,ytr,yts=train_test_split(ip,op,test_size=0.2, random_state=3)
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
sc=StandardScaler()
sc.fit(xtr)
xtr=sc.transform(xtr)
xts=sc.transform(xts)
from sklearn.ensemble import RandomForestClassifier
count = 0
for i in range(1, 100):
 clf = RandomForestClassifier(n_estimators=i, criterion='entropy', max_depth=4, random_state=1)
 clf.fit(xtr, ytr)
 yp6 = clf.predict(xts)
 accuracy=metrics.accuracy_score(yts,yp6)
 count = count +1
 if accuracy > 0.86:
   print(count)

   print(accuracy)
```
